#### 1. Изучите методы для сбора данных курсов валют в [API Центрального банка Российской Федерации](https://www.cbr.ru/development/SXML/) . Используя запросы к API, приведите значения заработной платы к рублям одним из вариантов: 1) простой вариант - на текущую дату  2) вариант посложнее - на дату создания (created).

```python
# ВАШ КОД ЗДЕСЬ 
import requests
from bs4 import BeautifulSoup
from re import sub
#Смотрим, какие валюты вообще есть в датасете
vacancy_read['currency'].value_counts()
# Кроме рублей, USD и EUR
# Переменная url страницы
url = 'http://www.cbr.ru/scripts/XML_daily.asp?' 
# Отправляем запрос к заданному url
xlm_inf = requests.get(url)
# Переводим полученный ответ в текст и размечаем его по тегам
soup = BeautifulSoup(xlm_inf.text, features="xml")
#доллар США
usd=soup.find(ID="R01235")
usd=usd.find('Value').text
#usd=float(usd)
#евро
eur=soup.find(ID="R01239")
eur=eur.find('Value').text
#eur=float(sub(',', '.', eur)) #Основа кода для пересчета
```

```python
#Пример преобразования
vacancy_read.loc[(vacancy_read['currency'] == 'USD'), 'salary_from'] *= float(sub(',', '.', eur))
vacancy_read[vacancy_read.currency=='USD']
```

```python
#пишем функцию...
#переименовать переменные!!! и перепроверить на свежую голову
# +преобразовать столбец валют
a='USD'
b='EUR'
c='salary_from'
d='salary_to'
def f(a, c, eur):
    vacancy_read.loc[(vacancy_read['currency'] == a), c] *= float(sub(',', '.', eur))
    return(vacancy_read[vacancy_read.currency==a])
f(b, c, eur)
f(b, d, eur)

f(a, c, usd)
f(a, d, usd)


```

#### 3. Сделайте рейтинг (топ-30) навыков по всем вакансиям.

```python

#удаляем строки с пустыми значениями в skills
vacancy_read['skills'].astype(bool)
vacancy_read = vacancy_read[vacancy_read['skills'].astype(bool)] 

#разделяем навыки для удобства
vacancy_read = vacancy_read["skills"].str.split(";").explode().str.strip().to_frame().combine_first(vacancy_read)

#выводим топ-30 навыков
print("Топ-30 навыков по всем вакансиям:")
print()
#находим топ-30 навыков
value_counts = vacancy_read['skills'].value_counts().head(30)

# преобразование в df и присвоение новых имен колонкам
df_value_counts = pd.DataFrame(value_counts)
df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['', 'Количество'] 
#Индекс с 1
df_value_counts.index += 1
#Удаляем столбец с количеством повторений, так как он не нужен
df_value_counts=df_value_counts.drop(columns = ['Количество'], axis = 1)
print(df_value_counts)





```
